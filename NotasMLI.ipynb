# Decision tree:

## Otros algoritmos de clasificacion:

### Decision tree:

Es un algoritmo mayormente utilizado para clasificaciones. Se llega a una conclusion y en base a esa conclusion, se predice la varible categorica. Es una serie de preguntas, un cuestionario.

Es un arbol y cada nodo es una pregunta, y en base a la respuesta, se llega a otro nodo hasta llegar a una hoja (un nodo sin hijos).

No utilza gradient descent sino que utiliza ID3, que se basa en la entropia.  A mayor entropia, menos informacion y mayor incertidumbre. La construccion de un arbol de decision es bastante complejo, por lo que solo se vera con scikit learn.

**VENTAJAS**

* Sumamente facil de explicar como se llego a una conclusion(prediccion)
* En scikit learn es facil imprimir una representacion grafica del arbol.

**DESVENTAJAS**

* Es muy probable que presente overfitting.
* El algortimo ID3 es muy costoso computacionalmente 
* En dataset muy grandes no funciona bien, dado que es costoso computacionalmente.



Para realizar predicciones recurrimos al comando *predict* de scikit learn

```python
# Ejemplo:  Utilizacion de scikit learn:

from sklearn.datasets import load_iris #Importar el dataset a utilizar.
from sklearn import tree #Importar la opcion para crear el algorimto de Decision Trees
import graphviz #Esta libreria se utilizara para graficar el arbol de decision
iris = load_iris() #Carga del dataset
clf = tree.DecisionTreeClassifier() #Crear del objeto (clasificador, predictor para otro tipo de tareas.)
clf = clf.fit(iris.data, iris.target) #Entrenamiento con scikit learn


# Opciones para grafica del contenido.
dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=iris.feature_names,  
                     class_names=iris.target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

# Regularizacion: 

El overfitting y el underfitting afectan tanto a modelos de clasificacion como de regresion. Lo primero que se necesita hacer es un proceso de diagnostico/debugging.

k folds validation:

En lugar de dividirlos en dos, hay que dividirlos en mas (k partes). IID -> Datos igualmente distribuidos. 
Se busca encontrar parametros que aproximar lo mejor posible a los parametros poblacionales. **SE BUSCA GENERALIZACION**  Un modelo que no generaliza es un modelo que solo memoriza.

Los datos de prueba y de entrenamiento son dos *dataset* distintos de una poblacion.

Los dos problemas qe buscamos combatir, son el overfitting y el underfitting. **NOTA: UN MODELO PUEDE TENER AMBAS!!**

### Overfitting:

* Si el modelo se comporta bien en el entrenamiento pero no en el de prueba -----> OVERFITTING. Tiende a ocurrir cuando tenemos modelos con demasiadas caracteristicas (features), muchas neuronas por capas, o mucha capacity. En estos casos **la capacidad (grado del polinomio), es mayor que el requerido**

* En cuanto a porblemas de clasificacion, la forntera de decision tiene formas muy complejas. Lo que se busca tambien es que el error en la data de prueba sea menor o igual al error en la data de entrenamiento. El error de generalizaciones **Es el error en el set de prueba** y nos indica **que tan bien nuestro modelo generaliza**

* El overfitting se conoce como error debido a varianza. cuando se hace inferencia estadisitca, el error se disminuye al aumentar el numero de la muestra. 


### Underfitting: 

* Un error grande tanto en la data de train como en la de test.

* Muy poca capacidad (grado del polinomio) para la tarea que se quiere resolver.

* Modelos con muy pocas caracteristicas.

* Muy pocas capas, o muy pocas neuronas por capa.

* Error debido a sesgo.

**CRITERIO** A mayor capacidad, es mas probable que se tenga overfitting, muy poca capacidad indica Underfitting.

### Debugging:

* Si el error es alto en ambos datos (test y train) tenemos underfitting.

* Si el error es alto unicamente en los datos (test) tenemos overfitting.

### Regularizacion:

Cualquier tecnica que tenga como proposito reducir el overfitting.

* Lasso (L1): Se agrega un tremino de penalizacion a la funcion de costo. Esta tecnica tiende a generar pesos (o parametros) pequenos. Se puede utilizar no solo para regulrizacion, sino para qreducir variables. La ventaja de Lasso es que puede utilizarse no solo para regularizacion sino para regresion de variables.
